Data Engineering
1. BAsed on Tenure value categorize Customers as New Customer, RegularCustomer , Loyal Customer

```python
bins=[-1,12,36,float('inf')]
labels=[ "New Customer", "Regular Customer", "Loyal Customer"]
df["Customer Segment"]=pd.cut(df['Tenure Months'],bins,labels=labels)
```

```python
df['Customer Segment'].value_counts()
```

based on Monthly Charges.

```python
# Define thresholds for Monthly Charges (e.g., $0-35 Low, $35-70 Medium, $70-100 High, $100+ Premium)
bins = [18,40,65,90, float('inf')]
labels = ["Low", "Medium", "High", "Premium"]

# Segment customers based on Monthly Charges
df["Revenue Segment"] = pd.cut(df['Monthly Charges'], bins=bins, labels=labels)
```

```python
df['Revenue Segment'].value_counts()
```

```python
#Service Adoption Score
services = [
    "Online Security",
    "Online Backup",
    "Device Protection",
    "Tech Support",
    "Streaming TV",
    "Streaming Movies"
]

df["Service Adoption Score"] = (
    (df[services] == "Yes")
    .sum(axis=1)
)

df["Service Adoption Score"].value_counts()
```

2219 customers don't use any of the six additional services.
284 customers use all six services.

```python
#Churn Score(highest are the high risk customers)
df.columns
```

```python
df['Churn Score'].describe()
```

```python
bins=[4,30,60,80,100]
labels=['Low Risk','Mediun Risk','High risk','Critical Risk']
df['Risk Category']=pd.cut(df['Churn Score'],bins=bins,labels=labels,include_lowest=True)
```
