```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"Telco_customer_churn.xlsx")
```

```python
print("Shape of dataset: ",df.shape)
df.info()
```

```python
#Check missing values
missing_values=df.isnull().sum()
print("Missing values : ",missing_values)
```

Churn Reason has 5174 missing values . Only 1869 values are present

```python
Duplicate_IDs=df['CustomerID'].duplicated().sum()
print("Duplicate customer id :",Duplicate_IDs)
```

No duplicates

```python
#Check invalid monhtly charges
invalid_monthly_charge=df[df['Monthly Charges'] < 0]
print("Invalid monthly charge",len(invalid_monthly_charge))
```

```python
df['Monthly Charges'] == 0
```

No invalid monthly charges

```python
#Check Invalid tenure months
Invalid_tenure_months=df[df['Tenure Months'] < 0]
print("Invalid tenure months ",len(Invalid_tenure_months))
```

No invalid tenure months

```python
#Check invalid cltv(Customer Lifetime Value)
invalid_cltv = df[df['CLTV'] < 0]

print("Invalid CLTV:", len(invalid_cltv))

```

No invalid cltv

```python
#Check if blank data in total charges column
blank_charges = df[df['Total Charges'].isna()]
print("Rows with blank Total Charges:",blank_charges)
```

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f"\n{col}")
    print(df[col].unique())
```

```python
quality_report = pd.DataFrame({
    'Issue': [
        'missing_values',
        'Duplicate_IDs',
        'invalid_monthly_charge',
        'Invalid_tenure_months',
        'invalid_cltv',
        'blank_charges'
    ],
    'Count': [
        df.isnull().sum().sum(),
        df['CustomerID'].duplicated().sum(),
        len(invalid_monthly_charge),
        len(Invalid_tenure_months),
        len(invalid_cltv),
        len(blank_charges)
    ]
})

quality_report
```

5174 missing values are usually from the Churn Reason column because customers who haven't churned don't have a churn reason.

```python
quality_report.to_excel("Data_Quality_Report.xlsx")
```
